```python

import pandas as pd
import numpy as np
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder, StandardScaler
from scipy.stats import multivariate_normal

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^\w\s]", " ", text)
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

with open("train_data.txt", encoding='utf-8') as f:
    train_lines = f.readlines()
train_data = [line.strip().split(" ::: ") for line in train_lines if len(line.strip().split(" ::: ")) == 4]
df_train = pd.DataFrame(train_data, columns=["ID", "TITLE", "GENRE", "DESCRIPTION"])
df_train["clean_desc"] = df_train["DESCRIPTION"].apply(clean_text)

with open("test_data.txt", encoding='utf-8') as f:
    test_lines = f.readlines()
test_data = [line.strip().split(" ::: ") for line in test_lines if len(line.strip().split(" ::: ")) == 3]
df_test = pd.DataFrame(test_data, columns=["ID", "TITLE", "DESCRIPTION"])
df_test["clean_desc"] = df_test["DESCRIPTION"].apply(clean_text)

with open("test_data_solution.txt", encoding='utf-8') as f:
    sol_lines = f.readlines()
sol_data = [line.strip().split(" ::: ") for line in sol_lines if len(line.strip().split(" ::: ")) == 4]
df_solution = pd.DataFrame(sol_data, columns=["ID", "TITLE", "GENRE", "DESCRIPTION"])
y_test = LabelEncoder().fit(df_train["GENRE"]).transform(df_solution["GENRE"])

vectorizer_full = TfidfVectorizer(max_features=5000)
X_raw = vectorizer_full.fit_transform(df_train["clean_desc"]).toarray()
X_raw_test = vectorizer_full.transform(df_test["clean_desc"]).toarray()

vectorizer_small = TfidfVectorizer(max_features=2000)
X_small = vectorizer_small.fit_transform(df_train["clean_desc"]).toarray()
X_small_test = vectorizer_small.transform(df_test["clean_desc"]).toarray()

scaler = StandardScaler()
X_small_std = scaler.fit_transform(X_small)
X_small_test_std = scaler.transform(X_small_test)

pca = PCA(n_components=100)
X_train_pca = pca.fit_transform(X_small_std)
X_test_pca = pca.transform(X_small_test_std)

lda = LDA(n_components=min(len(np.unique(y_test)) - 1, 30))
y_train = LabelEncoder().fit_transform(df_train["GENRE"])
X_train_lda = lda.fit_transform(X_small_std, y_train)
X_test_lda = lda.transform(X_small_test_std)

class BayesianClassifier:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.class_priors = {}
        self.class_means = {}
        self.class_covs = {}
        for c in self.classes:
            X_c = X[y == c]
            self.class_priors[c] = X_c.shape[0] / X.shape[0]
            self.class_means[c] = np.mean(X_c, axis=0)
            cov = np.cov(X_c, rowvar=False)
            self.class_covs[c] = cov + 1e-3 * np.identity(X.shape[1])

    def predict(self, X):
        preds = []
        for x in X:
            posteriors = []
            for c in self.classes:
                prior = np.log(self.class_priors[c])
                likelihood = multivariate_normal.logpdf(x, mean=self.class_means[c], cov=self.class_covs[c], allow_singular=True)
                posteriors.append(prior + likelihood)
            preds.append(np.argmax(posteriors))
        return np.array(preds)

acc_dict = {}
cv_dict = {}

def evaluate_model(Xtr, Xte, title):
    model = BayesianClassifier()
    model.fit(Xtr, y_train)
    y_pred = model.predict(Xte)

    acc = accuracy_score(y_test, y_pred)
    acc_dict[title] = acc

    print(f"\n[{title}] Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{title} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

evaluate_model(X_train_pca, X_test_pca, "Bayesian (PCA)")
evaluate_model(X_train_lda, X_test_lda, "Bayesian (LDA)")

print("\n=== Final Accuracy on Test Set ===")
for model, score in acc_dict.items():
    print(f"{model}: {score:.4f}")
```
